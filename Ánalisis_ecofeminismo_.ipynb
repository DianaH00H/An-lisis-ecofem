```python
# Importar librerías
import pandas as pd

# Cargar los archivos
file_path_mine = '/content/9_Diana_BD_Evidencia(Diana Laura (8)).csv'
file_path_zero = '/content/0_EH1021_BD_Evidencia1(EH1021).csv'

# Leer los datos
data_mine = pd.read_csv(file_path_mine, encoding='latin1')
data_zero = pd.read_csv(file_path_zero, encoding='latin1')

# Mostrar las primeras filas de ambos archivos como DataFrames
print("Primeras filas del archivo de mis encuestados:")
display(data_mine.head())  # Vista organizada

print("\nPrimeras filas del archivo cero:")
display(data_zero.head())  # Vista organizada

# Seleccionar tres filas aleatorias del archivo "cero"
random_rows = data_zero.sample(n=3, random_state=42)

# Mostrar las filas seleccionadas como DataFrame
print("\nTres filas aleatorias seleccionadas del archivo cero:")
display(random_rows)

# Combinar estas filas con el archivo "mío"
combined_data = pd.concat([data_mine, random_rows], ignore_index=True)

# Mostrar las últimas filas del archivo combinado como DataFrame
print("\nÚltimas filas del archivo combinado:")
display(combined_data.tail())  # Vista organizada

# Guardar el archivo combinado
combined_data.to_csv('archivo_combined.csv', index=False, encoding='latin1')

# Confirmar éxito
print("\nSe combinaron los datos y el archivo se guardó como 'archivo_combined.csv'")
```

```python
# Estandarizar la columna "Género"
def estandarizar_genero(genero):
    genero = str(genero).strip().lower()  # Convertir a minúsculas y eliminar espacios
    if genero in ['mujer', 'f']:
        return 'Mujer'
    elif genero in ['hombre', 'm']:
        return 'Hombre'
    elif genero in ['prefiero no contestar', 'n/a', 'na']:
        return 'Prefiero no contestar'
    else:
        return 'Prefiero no contestar'  # Valor predeterminado para datos inconsistentes

combined_data['Género'] = combined_data['Género'].apply(estandarizar_genero)

# Estandarizar la columna "Semestre"
def estandarizar_semestre(semestre):
    semestre = str(semestre).strip().lower()  # Convertir a minúsculas y eliminar espacios
    if 'primero' in semestre or '1' in semestre:
        return '1ro'
    elif 'segundo' in semestre or '2' in semestre:
        return '2do'
    elif 'tercero' in semestre or '3' in semestre:
        return '3ro'
    elif 'cuarto' in semestre or '4' in semestre:
        return '4to'
    elif 'quinto' in semestre or '5' in semestre:
        return '5to'
    elif 'sexto' in semestre or '6' in semestre:
        return '6to'
    else:
        return 'No especificado'  # Valor predeterminado para datos inconsistentes

combined_data['Semestre'] = combined_data['Semestre'].apply(estandarizar_semestre)

# Verificar los resultados
print("\nDatos estandarizados:")
display(combined_data[['Género', 'Semestre']].drop_duplicates())
```

```python
# Crear un DataFrame con las columnas demográficas incluyendo las estandarizadas
datos_demograficos = combined_data[['Edad', 'Género', 'Semestre', 'Carrera']]

# Mostrar el DataFrame combinado con las columnas demográficas
print("\nDatos demográficos completos:")
display(datos_demograficos)
```

```python
# Crear un nuevo DataFrame con las columnas seleccionadas
nivel_conocimiento = combined_data[
    [
        "¿Has cursado materias en prepa o profesional enfocadas al medioambiente?",
        "Tengo conocimiento sobre cómo el cambio climático afecta de manera desigual a las mujeres en comunidades vulnerables.",
        "¿Con qué frecuencia consultas fuentes confiables (revistas científicas, organizaciones ambientales) para obtener información sobre sostenibilidad?"
    ]
]

# Mostrar el nuevo DataFrame
print("\nNivel de conocimiento sobre la crisis climática:")
display(nivel_conocimiento)
```

```python
# Filtrar las filas con los valores específicos en la segunda columna
nivel_conocimiento_filtrado = nivel_conocimiento[
    nivel_conocimiento[
        "Tengo conocimiento sobre cómo el cambio climático afecta de manera desigual a las mujeres en comunidades vulnerables."
    ].isin(["Moderadamente de acuerdo", "Totalmente de acuerdo"])
]

# Mostrar el DataFrame filtrado
print("\nNivel de conocimiento sobre la crisis climática (filtrado):")
display(nivel_conocimiento_filtrado)
```

```python
from matplotlib import pyplot as plt
import seaborn as sns
import pandas as pd
plt.subplots(figsize=(8, 8))
df_2dhist = pd.DataFrame({
    x_label: grp['¿Con qué frecuencia consultas fuentes confiables (revistas científicas, organizaciones ambientales) para obtener información sobre sostenibilidad?'].value_counts()
    for x_label, grp in nivel_conocimiento_filtrado.groupby('Tengo conocimiento sobre cómo el cambio climático afecta de manera desigual a las mujeres en comunidades vulnerables.')
})
sns.heatmap(df_2dhist, cmap='viridis')
plt.xlabel('Tengo conocimiento sobre cómo el cambio climático afecta de manera desigual')
_ = plt.ylabel('frecuencia de consultas de fuentes confiables sobre sostenibilidad?')
```

```python
# Crear un nuevo DataFrame con las columnas seleccionadas
influencia_y_medios = combined_data[
    [
        "¿Qué medios de comunicación consideras más confiables para informarte sobre problemas ambientales?",
        "¿Quién o qué ha sido lo más influyente en la formación de tus puntos de vista sobre la sostenibilidad?"
    ]
]

# Mostrar el nuevo DataFrame
print("\nMedios de comunicación e influencia en sostenibilidad:")
display(influencia_y_medios)
```

```python
# Filtrar el DataFrame "influencia_y_medios" basado en los índices de "nivel_conocimiento_filtrado"
influencia_y_medios_filtrado = influencia_y_medios.loc[nivel_conocimiento_filtrado.index]

# Mostrar el DataFrame filtrado
print("\nMedios de comunicación e influencia en sostenibilidad (filtrado):")
display(influencia_y_medios_filtrado)
```

```python
# @title ¿Qué medios de comunicación consideras más confiables para informarte sobre problemas ambientales?

from matplotlib import pyplot as plt
import seaborn as sns
influencia_y_medios_filtrado.groupby('¿Qué medios de comunicación consideras más confiables para informarte sobre problemas ambientales?').size().plot(kind='barh', color=sns.palettes.mpl_palette('Dark2'))
plt.gca().spines[['top', 'right',]].set_visible(False)
```
